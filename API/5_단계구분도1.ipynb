# 단계구분도(choropleth)

```python
# 서울시 구별 CCTV 인구
```

```python
import json, folium
import pandas as pd
```

```python
df = pd.read_csv('test_data/서울시 구별 CCTV 인구 현황.csv', index_col='구별')
df
```

```python
geo_data = json.load(open('test_data/seoul_geo_simple.json', encoding='utf-8'))
```

```python
geo_data
```

```python
geo_data.keys()
```

```python
geo_data['features'][0]
```

```python
center = [37.581, 126.986]
map = folium.Map(location=center, zoom_start=11)
folium.Choropleth(
    geo_data=geo_data,
    data=df.인구수, # 단계구분도에서 보여줄 데이터 
    columns=[df.index, df.인구수],
    key_on='feature.id',
    fill_color='BuPu',
    legend_name='노령 인구수'
).add_to(map)
map
```

```python
# 자치구 이름쓰기
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            # print(geo)
            gu_dict[gu['id']] = [np.mean(geo[:, 1]), np.mean(geo[:,0])]
            # print(gu_dict)
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data)
```

```python
gu_dict
```

```python
for gu_name in df.index:
    folium.map.Marker( 
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80, 20), icon_anchor=(20,0),
                              html=f'{gu_name}')
    ).add_to(map)
map
```
