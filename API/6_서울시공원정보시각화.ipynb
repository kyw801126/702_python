# 서울시 공원 시각화

```python
import pandas as pd
import numpy as np
import folium
```

```python
df = pd.read_csv('test_data/서울시 주요 공원현황.csv', encoding='euc-kr')
df.head(2)
```

```python
df.info()
```

```python
df = df[['공원명', '면적', '공원주소', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
df.columns = ['공원명', '면적', '주소', '경도', '위도']
df
```

```python
df.info()
```

```python
df.dropna(subset=['면적', '경도'], inplace=True)
df.isna().sum()
```

```python
# 면적에 따라 공원크기를 지도에 표시
df.면적.head(10)
```

```python
df.면적.tail(10)
```

```python
# 6행의 면적 컬럼은 면적 정보가 아닌 공원개요 정보로 생각되서 삭제
df.면적[6]
df.drop([6], inplace=True)
```

```python
import re
df['면적'] = df.면적.apply(lambda x: re.split('[㎡m²㎥m]', x)[0])
```

```python
df['면적'].head()
```

```python
# https://codefile.io/f/ohdaY7mKYk
```

```python
df['면적'] = df['면적'].str.replace('총 ', "")
df['면적'].unique()
```

```python
df['면적'] = df.면적.astype(float).astype(int)
```

```python
# 소형 < 100,000, 중형 < 100,0000, 대형20,000,000
criteria = [0, 100000, 1000000, 20000000]
labels = ['소형', '중형', '대형']
size_info = [3, 7, 15]
df['분류'] = pd.cut(df.면적, criteria, labels=labels)
df['크기'] = pd.cut(df.면적, criteria, labels=size_info) # 지도그릴때 면적을 원크기로 표현해주기 위함
```

```python
df.head()
```

```python
type(df.크기[0])
```

```python
df.to_csv('test_data/서울공원요약.csv', index=False)
```

# 공원시각화 

```python
park = folium.Map([37.55, 126.98], zoom_start=11)
for i in df.index:
    folium.CircleMarker(
        location=[df.위도[i], df.경도[i]],
        radius=int(df.크기[i]),
        tooltip=df.공원명[i],
        popup=folium.Popup(df.주소[i], max_width=200),
        color='crimson', fill_color='crimson'
    ).add_to(park)
title_html = '<h3 align="center" style="font-size:20px"> 서울시 공원 현황 </h3>'
park.get_root().html.add_child(folium.Element(title_html))
park
```
