# 무신사 크롤링

```python
import requests
from bs4 import BeautifulSoup
import re
# 무신사 사이트에서 아무거나 카테고리 아무거나 들어간다음 중분류 니트/스웨터 삭제 후에
# 무신사 추천을 누르고 오른쪽에 1페이지를 누르면됨
```

```python
# 1페이지 내용만 확인하기
url = f"https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page=1&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure="
respons = requests.get(url)
soup = BeautifulSoup(respons.text, 'html.parser')
```

```python
soup
```

```python
# 상품정보가 ul 테그 중에서 id가 searchList이면서 li테그에 있음
items = soup.select('ul#searchList > li')
items
```

```python
# 브랜드 명과 상품명 가져오기
print(items[3].select_one('.article_info p.item_title a').text.strip())
print(items[3].select_one('.article_info p.list_info a').attrs['title'])
```

```python
#스플릿으로 찾아서 0, 또는 -1 로 찾는 방법
price = items[0].select_one('.article_info p.price').text.strip().split(" ")
price
```

```python
# 정규표현식으로 찾는 방법
price = re.findall('[0-9,]*원', items[15].select_one('.article_info p.price').text.strip())
price
```

```python
if len(re.findall('[0-9,]*원', items[0].select_one('.article_info p.price').text.strip())) == 1:
    price = re.findall('[0-9,]*원', items[0].select_one('.article_info p.price').text.strip())[0]
    print(price)
else:
    price = re.findall('[0-9,]*원', items[0].select_one('.article_info p.price').text.strip())[1] 
    print(price)
```

```python
for i in range(1, 2):
    print(f'Page ==========, {i}')
    url = f"https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure="
    respons = requests.get(url)
    soup = BeautifulSoup(respons.text, 'html.parser')

    items = soup.select('ul#searchList > li')
    for item in items:
        brand = item.select_one('.article_info p.item_title a').text.strip()
        product = item.select_one('.article_info p.list_info a').attrs['title']

        if len(re.findall('[0-9,]*원', item.select_one('.article_info p.price').text.strip())) == 1:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text.strip())[0]
            
        else:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text.strip())[1]
        print("브랜드", brand)
        print("제품명", product)
        print("가격", price)
```
