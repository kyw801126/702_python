```python
import requests
import pandas as pd 
from bs4 import BeautifulSoup
from urllib.parse import quote
```

```python
baseurl = 'https://www.siksinhot.com/search'
url = f'{baseurl}?keywords={quote("신촌역")}' # %EC%8B%A0%EC%B4%8C%EC%97%AD
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
# soup = BeautifulSoup(res.text, 'lxml')
```

```python
lis = soup.select('.localFood_list > li')
print(len(lis))
```

```python
li = lis[0]
li
```

```python
atag = li.select_one('figcaption > a')
name = atag.select_one('h2').get_text().strip()
name
```

```python
menu = li.select('.cate > a')[-1].get_text().strip()
menu
```

```python
score = atag.select_one('.score').get_text().strip()
score
```

```python
img = li.select_one('figure > a > img').get('alt').split(',')[-1]
img 
```

```python
sub_href = atag.get('href')
# atag['href'] 또는 atag.attrs('href')
sub_href
```

```python
# mini_soup
mini_res = requests.get(sub_href)
mini_soup = BeautifulSoup(mini_res.text, 'html.parser')
mini_soup
```

```python
info = mini_soup.select('.pc_only > td')
info
#전화번호는 자바스크립트로 사이트가 개편되서 가져올수 없음
```

```python
data = [] 
for li in lis:
    atag = li.select_one('figcaption > a')
    name = atag.select_one('h2').get_text().strip()
    menu = li.select('.cate > a')[-1].get_text().strip()
    score = atag.select_one('.score').get_text().strip()
    addr = li.select_one('figure > a > img').get('alt').split(',')[-1]
    data.append({'업소명' : name, '평점' : score, '메뉴' : menu, 
                 '주소': addr})
```

```python
df = pd.DataFrame(data)
df.head()
df.to_csv('./data/식신신촌역.csv')
```
