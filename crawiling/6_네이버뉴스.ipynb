# 네이버 뉴스 크롤링

```python
import requests
from bs4 import BeautifulSoup
url = 'https://news.naver.com/main/ranking/popularDay.naver'
headers = { "User-Agent" : "Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36"}
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
newslist = soup.select(".rankingnews_list")
len(newslist)
```

```python
news = newslist[0]
news
```

```python
print(news.find_all('li'))
print(len(news.find_all('li')))
```

```python
lis = news.find_all('li')
lis[0]
```

```python
news_rank = lis[0].select_one(".list_ranking_num").text
news_rank 
```

```python
news_title = lis[0].select_one(".list_title")
news_title_text = news_title.text
```

```python
news_link = news_title.get('href')
news_link
```

```python
newsData = []
for news in newslist[:16]:
    lis = news.find_all('li')
    print(lis)
    for li in ls:
        news_rank = li.select_one(".list_ranking_num").text
        news_title = li.select_one(".list_title")
        news_title_text = news_title.text
        news_link = news_title.get('href')
        newsData.append({"ranking" : news_rank,
                         "title" : news_title_text,
                        "link": news_link })
```

```python
len(newsData)
```

```python
newsData
```

```python
# 뉴스 작성 시간
news_url = newsData[0]['link']
news_url
```

```python
mini_res = requests.get(news_url, headers=headers)
mini_soup= BeautifulSoup(mini_res.text, 'html.parser')
news_time = mini_soup.select_one(".media_end_head_info_datestamp").select_one(".media_end_head_info_datestamp_time").attrs['data-date-time']
print(news_time)
```

```python
news_contents = mini_soup.select_one("#newsct_article").text.replace("\n","").replace("\t","")
news_contents
```

```python
for news in newsData:
    news_url = news['link']
    mini_res = requests.get(news_url, headers=headers)
    mini_soup= BeautifulSoup(mini_res.text, 'html.parser')
    news_time = mini_soup.select_one(".media_end_head_info_datestamp").select_one(".media_end_head_info_datestamp_time").attrs['data-date-time']
    news_contents = mini_soup.select_one("#newsct_article").text.replace("\n","").replace("\t","")
    news['time'] = news_time
    news['contents'] = news_contents
```

```python
import pandas as pd
df = pd.DataFrame(newsData)
df
```

```python
df.to_csv("네이버뉴스수집결과.csv")
```
