```python
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

* Stop word 비교

```python
sw_nltk = stopwords.words('english')
print(sw_nltk)
print(len(sw_nltk))
```

```python
# wordcloud에서 제공하는 불용어
sw_wc = STOPWORDS
print(sw_wc)
print(len(sw_wc))
```

### 1.Alice in Wonderland

```python
with open('Alice.txt') as file:
    text = file.read()
```

```python
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=sw_wc
).generate(text)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 사용빈도 Top 10
count = 0
for key, val in wc.words_.items():
    print(f'{key}: \t{val}')
    count += 1
    if count >= 10:
        break
```

```python
# stopwords에 said 추가
sw_wc.add('said')
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=sw_wc
).generate(text)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
count = 0
for key, val in wc.words_.items():
    print(f'{key} : \t{val}')
    count += 1
    if count >= 10:
        break
```

* 배경 그림위 표시

```python
mask = np.array(Image.open('Alice_mask.png'))
plt.figure(figsize=(6,10))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=sw_wc,
    mask=mask
).generate(text)
plt.figure(figsize=(6, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
