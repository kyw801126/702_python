## 승차/하차 테이블 분리

### 승차 테이블 만들기

```python
df.head()
```

```python
# 공통 컬럼
df1 = df.iloc[: , :3]
# 승차 컬럼만 가져오기
df2 = df.iloc[: , 3::2]
df2
```

```python
# 공통컬럼과 승차컬럼 연결하기
df_in = pd.concat([df1, df2], axis=1)
df_in
```

## 하차 테이블 만들기

```python
# 공통 컬럼

# 하차 컬럼만 가져오기
df3 = df.iloc[: , 4::2]
```

```python
# 공통컬럼과 하차컬럼 연결하기
df_out = pd.concat([df1, df3], axis=1)
df_out
```

```python
df_in_2108_agg.loc['sum'] = df_in_2108_agg.apply('sum', axis=0)
df_in_2108_agg
```
