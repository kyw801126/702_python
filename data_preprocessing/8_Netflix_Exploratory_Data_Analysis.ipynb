```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['country']
```

```python
df['country'].value_counts()
```

```python
df.isna().sum()
```

```python
df['cast'].replace(np.nan, 'Missing', inplace=True)
df['director'].replace(np.nan, 'Missing', inplace=True)
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['date_added']
```

```python
date_str = ['2023-09-12', '2023.9.13', '2023, 9,14', '9/14/2023']
pd.to_datetime(date_str)
# 주피터는 format 옵션을 추가 해야함 pd.to_datetime(date_str, format='mixed')
```

```python
df['month_added'] = df['date_added'].dt.month
df['month_name_added'] = df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
df.tail()
```

```python
df['country_1st'] = df['country'].apply(lambda x: x.split(',')[0])
df['country_1st'].tail()
```

```python
df['country_1st'].replace('United States', 'USA', inplace=True)
df['country_1st'].replace('United Kingdom', 'UK', inplace=True)
df['country_1st'].replace('South Korea', 'S. Korea', inplace=True)
df['country_1st'].tail()
```

```python
df['rating'].unique()
```

```python
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
```

```python
df['target_user'] = df['rating'].replace(ratings_ages)
df['target_user'].unique()
```

```python
df.tail()
```

```python
df['duration'].unique()
```

```python
df['season_count'] = df.apply(lambda x : x['duration'].split(' ')[0] if 'Season' in x['duration'] else "", axis=1)
df['season_count'].unique()
```

```python
df['season_count'].replace('', np.nan, inplace=True)
df['season_count'].unique()
```

```python
df['duration'] = df.apply(lambda x : x['duration'].split(' ')[0] if 'Season' not in x['duration'] else np.nan, axis=1)
df['duration'].unique()
```

```python
df.head()
```

```python
df = df.drop(columns=['show_id', 'description'])
df.head()
```

Change Dtypes

```python
df.dtypes
```

```python
df['type'].unique()
```

```python
df['target_user'].unique()
```

```python
df['type'] = pd.Categorical(df['type'])
df['target_user'] =  pd.Categorical(df['target_user'])
df['year_added'] = pd.to_numeric(df['year_added'])
df['duration'] = pd.to_numeric(df['duration'])
df['season_count'] = pd.to_numeric(df['season_count'])
df.dtypes
```

```python
df_tv = df[df['type'] == 'TV Show']
```

```python
df_movie = df[df['type'] == 'Movie']
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=df, palette='husl')
sns.despine(left=True)
plt.title('콘텐츠 타입비중', fontsize=14, fontweight='bold', position=(0,0))
plt.show()
```

```python
type_count = df.groupby(['type'])['type'].count()
length = len(df)
result =  (type_count / length).round(2)
result
```

```python
result.T
```

```python
type_ration = pd.DataFrame(result).T
type_ration
```

```python
labels = ['Movie', 'Tv show']
plt.figure(figsize=(6,4))
plt.pie(type_count/length, labels=labels, autopct='%1.2f%%', textprops={'fontsize':10})
plt.title('콘텐츠타입비율', fontweight='bold', position=(0,0))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='month_added', hue='type', data=df, palette='husl')
sns.despine(left=True)
plt.title('월별 콘텐츠 업로드 수', fontsize=14, fontweight='bold', position=(0,0))
plt.show()
```

```python
# release_year와 type을 연도순으로 hue = type
plt.figure(figsize=(10,5))
sns.countplot(x='release_year', hue='type', data=df, palette='husl')
sns.despine(left=True)
plt.title('연도순 콘텐츠 개봉 or 방영수', fontsize=14, fontweight='bold', position=(0,0))
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='target_user', hue='type', data=df, palette='husl')
sns.despine(left=True)
plt.title('콘텐츠 타입별 타겟 유저수', fontsize=14, fontweight='bold', position=(0,0))
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='year_added', hue='type', data=df, palette='husl')
sns.despine(left=True)
plt.title('연도순 콘텐츠 타입별 업로드수', fontsize=14, fontweight='bold', position=(0,0))
plt.xticks(rotation=90)
plt.show()
```

```python
# 영화재생시간
df_duration_over0 = df[df['duration'] > 0]
plt.figure(figsize=(10,5))
sns.histplot(x='duration', data=df_duration_over0, kde=True, bins=30,
             binwidth=4, discrete=False)
#kernel density estimate
sns.despine(left=True)
plt.title('영화 재생시간 분포', fontsize=14, fontweight='bold', position=(0,0))
plt.xticks(rotation=90)
plt.show()
```

```python
# 시즌
df_season_count = df['season_count'].value_counts().reset_index()
df_season_count
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='index', y='season_count', data=df_season_count)
#kernel density estimate
sns.despine(left=True)
plt.title('Tv쇼 시즌 수', fontsize=14, fontweight='bold', position=(0,0))
plt.xticks(rotation=90)
plt.show()
```

```python
# 월별 컨텐츠 업로드수 히트맵
#구룹 묶기
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
               'September', 'October', 'November', 'December'][::-1]
df_bymonth = df.groupby('year_added')['month_name_added'].value_counts().unstack().fillna(0)[month_order].T
```

```python
df.groupby('year_added')['month_name_added'].value_counts()
```

```python
df.groupby('year_added')['month_name_added'].value_counts().unstack()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df_bymonth, linewidths=.5, cmap='YlGnBu')
plt.title('연도&월별 업로드수 히트맵', fontsize=14, fontweight='bold', position=(0,0))
plt.show()
```

```python
# 제작국가
top10_country = df.groupby('country_1st')['country'].agg('count').sort_values(ascending=False)[:10]
```

```python
top10_country
```

```python
colors = ['#f1f1f1' for _ in range(len(top10_country))]
colors[0] = '#E50914'
# ['#E50914' ,'#f1f1f1' ,'#f1f1f1' ,'#f1f1f1' ,'#f1f1f1' ,'#f1f1f1' ....]
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10_country.index, top10_country.values, width=0.8, linewidth=0.6, color=colors)
plt.title('Top10 제작국가', fontsize=14, fontweight='bold', position=(0,0))
sns.despine(top=True, right=True, left=True, bottom=False)
plt.show()
```
